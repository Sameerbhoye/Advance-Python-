```python
import numpy as np
import pandas as pd
import yfinance as yf
from keras.models import Sequential
from keras.layers import LSTM, Dense
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from multiprocessing import Process
import matplotlib.pyplot as plt
```

```python
# --- Function to train LSTM and forecast future prices ---
def lstm_stock_forecast(stock_symbol, future_days=30, epochs=10):
    print(f"ğŸ”¹ Training LSTM for {stock_symbol}...")

    # 1ï¸âƒ£ Fetch historical stock data
    df = yf.download(stock_symbol, start="2020-01-01", end="2024-12-31")
    if df.empty:
        print(f"No data found for {stock_symbol}")
        return

    data = df['Close'].values.reshape(-1, 1)

    # 2ï¸âƒ£ Scale data
    scaler = MinMaxScaler(feature_range=(0, 1))
    data_scaled = scaler.fit_transform(data)

    # 3ï¸âƒ£ Prepare sequences
    X, y = [], []
    for i in range(60, len(data_scaled)):
        X.append(data_scaled[i-60:i, 0])
        y.append(data_scaled[i, 0])
    X, y = np.array(X), np.array(y)
    X = np.reshape(X, (X.shape[0], X.shape[1], 1))

    # 4ï¸âƒ£ Split train/test
    split = int(len(X) * 0.8)
    X_train, X_test = X[:split], X[split:]
    y_train, y_test = y[:split], y[split:]

    # 5ï¸âƒ£ Build LSTM model
    model = Sequential([
        LSTM(50, return_sequences=False, input_shape=(X_train.shape[1], 1)),
        Dense(1)
    ])
    model.compile(optimizer='adam', loss='mean_squared_error')

    # 6ï¸âƒ£ Train model
    model.fit(X_train, y_train, epochs=epochs, batch_size=32, verbose=0)

    # 7ï¸âƒ£ Predict test data
    predicted_scaled = model.predict(X_test)
    predicted = scaler.inverse_transform(predicted_scaled)
    actual = scaler.inverse_transform(y_test.reshape(-1,1))

    rmse = np.sqrt(mean_squared_error(actual, predicted))
    print(f"{stock_symbol} RMSE on historical test data: {rmse:.4f}")

    # 8ï¸âƒ£ Forecast future prices
    last_60_days = data_scaled[-60:]
    future_predictions_scaled = []
    current_input = last_60_days.reshape(1,60,1)

    for _ in range(future_days):
        next_pred_scaled = model.predict(current_input)
        future_predictions_scaled.append(next_pred_scaled[0,0])
        next_pred_scaled = next_pred_scaled.reshape(1,1,1)  # fix dimension
        current_input = np.append(current_input[:,1:,:], next_pred_scaled, axis=1)

    future_predictions = scaler.inverse_transform(np.array(future_predictions_scaled).reshape(-1,1))

    # 9ï¸âƒ£ Plot results
    plt.figure(figsize=(10,5))
    plt.plot(np.arange(len(data)), data, label='Historical Price', color='blue')
    plt.plot(np.arange(split+60, len(data)), predicted, label='Predicted (Test)', color='orange')
    plt.plot(np.arange(len(data), len(data)+future_days), future_predictions, label=f'Forecast Next {future_days} Days', color='green')
    plt.title(f"{stock_symbol} - LSTM Stock Forecast")
    plt.xlabel("Time")
    plt.ylabel("Stock Price (USD)")
    plt.legend()
    plt.show()
```

```python
# --- List of stocks to forecast ---
stocks = ["AAPL", "MSFT", "GOOG"]

# --- Create processes for each stock (true parallelism) ---
processes = [Process(target=lstm_stock_forecast, args=(symbol,30,10)) for symbol in stocks]

# --- Start all processes ---
for p in processes:
    p.start()

# --- Wait for all processes to finish ---
for p in processes:
    p.join()
```
